```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/ge-2024-results1/GE_2024_Results (1).csv')
```

```python
df.head(20)
```

```python
df=df.drop(columns=['ID'])
df
```

```python
df=df.sort_index(axis = 0)
df
```

```python
df=df.dropna()
df
```

```python
df.loc[df["State"]=="Telangana"]["Total Votes"].sort_values(ascending=False).head(5)
```

```python
df.groupby("State")["Total Votes"].sum()
```

```python
df.groupby("State")["Total Votes"].sum()/df['Total Votes'].sum()*100
```

```python
df.groupby("Party")["Total Votes"].sum().sort_values(ascending=False).head(5)
```

```python
neww_df = df.loc[df['Result'] == 'Lost']
neww_df['% of Votes'] = neww_df['% of Votes'].astype(float) 
neww_df.sort_values('% of Votes', ascending = False,inplace = True)
neww_df.head(10)
```

```python
Nw_df = df.loc[df["Result"] == "Won"]
Nw_df = Nw_df.sort_values(by="% of Votes", ascending=False).head(5)
Nw_df.head(5)
labels = Nw_df['Candidate'].unique()
values = Nw_df['Total Votes'].unique()
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Candidate Vote Share")
plt.axis("equal")
plt.show()
```

```python
nw_df = (df.loc[df['State'] == 'Karnataka'])
parties = nw_df['Party'].unique()

nw_df.loc[:,'Postal Votes'] = nw_df['Postal Votes'].replace('-','0')
nw_df.loc[:,'Postal Votes'] = nw_df['Postal Votes'].astype(float)
votes = []
for i in parties:
     votes.append(nw_df.loc[df['Party'] == i, 'Postal Votes'].sum())

fig, ax = plt.subplots(figsize=(150,25))
plt.bar(parties, votes)
plt.yscale('log')
plt.xlabel('Parties')
plt.ylabel('Votes')
plt.title('postal votes received by the parties in Karnataka)')
plt.show()
```
